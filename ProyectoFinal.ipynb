```python
# Importando LabelEncoder para convertir valores categóricos en numéricos
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
# Iterando sobre todas las columnas del DataFrame
for col in oscars3.columns:
    # Si la columna es de tipo objeto (categórico)
    if oscars3[col].dtype == 'object':
        # Transforma la columna en valores numéricos
        encoder.fit_transform(list(oscars3[col].values))
        oscars3[col] = encoder.transform(oscars3[col].values)

# Mostrando las primeras 5 filas del DataFrame después de la codificación
oscars3.head()
```

```python
oscars3.info()
```

```python
selected_variables=['person','birthplace','religion','sexual_orientation', 'award' ,'movie','race_ethnicity']
selected_data=oscars3[selected_variables]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ordenando y mostrando las características basadas en su correlación con la columna 'awsrd'
selected_data.corr()['award'].sort_values(ascending=False)
```

```python
# Definiendo las características y la variable objetivo
X = selected_data[['birthplace','religion','sexual_orientation','award','race_ethnicity']]
Y = selected_data['person']
```

**PARTE 2:ENTRENAMIENTO DEL MODELO**

```python
from sklearn.model_selection import train_test_split
# Dividiendo el conjunto de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test= train_test_split(X,Y,test_size=0.25, random_state=1)
print(X_train)
print(y_train)

# Mostrando las características del conjunto de entrenamiento
```

```python
X_train.shape,y_train.shape
```

```python
#Escalamos los valores de X.
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
modelo1=LogisticRegression()
modelo1.fit(X_train,y_train)
```

```python
y_pred=modelo1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

exactitud= accuracy_score(y_test, y_pred)
print('Exactitud del modelo:',exactitud)
```

```python

from sklearn.metrics import confusion_matrix
```

```python
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de confusion:',matriz)
```

```python
from sklearn.metrics import precision_score

print("Precision del modelo : ",precision_score(y_test, y_pred,
                                           pos_label='positive',
                                           average='micro'))
```

```python
from sklearn.metrics import f1_score

print(" Puntaje F1 : ",f1_score(y_test, y_pred,
                                           pos_label='positive',
                                           average='micro'))
```

```python
from sklearn.metrics import recall_score
print(" Recall store : ",recall_score(y_test, y_pred,
                                           pos_label='positive',
                                           average='micro'))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, fmt= "d", cmap= "coolwarm",annot_kws={"size"})
plt.title('Matriz de confusión_LogisticRegresion')
plt.xlabel('Valor predecido')
plt.ylabel('Valor real')
plt.show()
```

**MODELO KNN**

```python
# Importamos el modelo KNN para regresión de la siguiente manera y lo asignamos a un objeto.
from sklearn.neighbors import KNeighborsClassifier
modelo_knn = KNeighborsClassifier(n_neighbors= 12)
```

```python
#Entrenamos el modelo
modelo_knn.fit(X_train, y_train)
```

```python
y_pred_knn = modelo_knn.predict(X_test)
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_knn
```

```python
print("La precisión del modelo usando KNN es de:", accuracy_score(y_test,y_pred_knn))
```

```python
print(" Puntaje F1 con KNN : ",f1_score(y_test, y_pred_knn,
                                           pos_label='positive',
                                           average='micro'))
```

```python
print("REcall store con KNN : ",recall_score(y_test, y_pred_knn,
                                           pos_label='positive',
                                           average='micro'))
```
